Our goal is to use data-driven insights to maximise streamas for taxi drivers in order to meet this need . Our research aims to determine whether payment methods have an impact on fare pricing by focusing on the relationship between payment type and fare amount.

```python
# using IQR to remove outlier becouse data is not distributed properly 
for col in ['fare_amount','trip_distance', 'duration']:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    IQR = q3 - q1

    lower_bound = q1 - 1.5*IQR
    upper_bound = q3 + 1.5*IQR

    df = df[(df[col]>=lower_bound) & (df[col]<=upper_bound)]
```

```python
# Capture the plot in the 'ax' variable
ax = data.plot(x='payment_type', kind='barh', stacked=True, color=['#FA643F','#FFBCAB','#F1F1F1','#FD9F9F'])

# Add percentage text
for p in ax.patches:
    width = p.get_width()
    # Only label bars that actually have a value > 0
    if width > 0:
        height = p.get_height()
        x, y = p.get_xy()
        ax.text(x + width/2, 
                y + height/2, 
                '{:.0f}%'.format(width), 
                horizontalalignment='center', 
                verticalalignment='center',fontsize=6)
plt.show()
```

```python
# Plot qq graph to see that Data distribution. If normaly distributed then it should be along the line or if not then it is not a normal distribution 
sm.qqplot(df['fare_amount'],line='45')
plt.show()
```

```python
card_sample = df[df['payment_type'] == 'Card']['fare_amount']
cash_sample = df[df['payment_type'] == 'Cash']['fare_amount']
```

### Applying T-test

```python
t_stats,p_value = st.ttest_ind(a=card_sample, b=cash_sample, equal_var=False)
print('T test Statistic',t_stats,'p_value',p_value)
```
